```python
import os
from dotenv import load_dotenv
from openai import OpenAI
```

```python
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set")

MODEL = "gpt-5"
openai = OpenAI()
```

```python
with open("recordings/", "r") as f: # Make sure to replace "recordings/" with the actual path to your file containing the recording URL
    recording_url = f.read().strip()
```

```python
system_prompt = f"""You are a helpful meeting assistant. You will be given a transcript of a meeting and your task is to summarize the key points and action items from the meeting. The transcript will be provided in the form of a .vtt file. Please read the transcript from the file and provide a summary of the meeting, including any action items that were discussed."""
```

```python
user_prompt = f"""Here is the transcript of the meeting: {recording_url}"""
```

```python
def summarize_meeting(system_prompt, user_prompt):
    response = openai.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )
    return response.choices[0].message.content
```

```python
summary = summarize_meeting(system_prompt, user_prompt) 
print(summary)
```
