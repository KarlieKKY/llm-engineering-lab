# Video file transcription

```python
import whisper
import tqdm
from pathlib import Path
```

```python
import torch
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using device:', device)

print(f"PyTorch: {torch.version.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
print(f"CUDA device: {torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'None'}")
print(f"Is CUDA active? {torch.cuda.is_available()}")
print(f"GPU Name: {torch.cuda.get_device_name(0)}")
```

```python
model = whisper.load_model("base", device=device)

base_path = r"" # Update this path to the directory containing your .mp4 files
mp4_files = list(Path(base_path).glob("*.mp4"))
# mp4_files = list(Path(base_path).glob("*.m4a"))

```

```python
print(f"Found {len(mp4_files)} MP4 files")
```

```python
def format_timestamp(seconds):
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    secs = int(seconds % 60)
    millis = int((seconds % 1) * 1000)
    return f"{hours:02d}:{minutes:02d}:{secs:02d}.{millis:03d}"
```

```python
for mp4 in tqdm.tqdm(mp4_files):
    result = model.transcribe(str(mp4)) 
    output_file = mp4.stem + ".vtt"  # Changes .mp4 to .vtt

    with open(output_file, "w") as f:
        f.write("WEBVTT\n\n")
        for segment in result["segments"]:
            start = segment["start"]
            end = segment["end"]
            text = segment["text"]
            f.write(f"{format_timestamp(start)} --> {format_timestamp(end)}\n")
            f.write(f"{text}\n\n")
    print(f"Saved: {output_file}")
```
